## Univariate Analysis

### Customer Rating Distribution

```python
sns.distplot(df['Rating'], kde=True)
plt.axvline(x=np.mean(df['Rating']), c='red', ls='--', label='Mean')
plt.axvline(x=np.percentile(df['Rating'], 25), c='green', ls='--', label='25-75th percentile')
plt.axvline(x=np.percentile(df['Rating'], 75), c='green', ls='--')
plt.legend()
```

Distribution for Customer Ratings looks uniform and there doesn't seem to be any left or right skew in the distribution.

### Distribution for other numerical features

```python
df.hist(figsize=(12,12))
```

- Unit price, quantity and rating look uniformly distributed
- Tax 5%, Total, cogs, gross income has right skew
- Gross Margin % is a constant value, so no distribution
- Tax 5%, Gross income, Total, cogs are highly correlated, so have similar distribution

## Dealing with Duplicate Rows and Missing Values

```python
df.duplicated().sum()
```

```python
df[df.duplicated()==True]
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.isna().sum()
```

df["Quantity"].fillna(df["Quantity"].mean(),inplace=True)
df["Unit price"].fillna(df["Unit price"].mean(),inplace=True)
df["Product line"].fillna(df["Product line"].mode()[0],inplace=True)
df["Customer type"].fillna(df["Customer type"].mode()[0],inplace=True)
df.isna().sum()

```python
sns.heatmap(df.isnull())
```
